```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imgBinQuan(img):
    # 
    nr,nc = img.shape[:2]
    # 
    new_img = np.zeros((nr, nc, 3), np.uint8)
    for i in range(nr):
        for j in range(nc):
            for k in range(3):  # RGB                
                if img[i, j][k] < 128:
                    gray = 0
                else:
                    gray = 255
                new_img[i, j][k] = np.uint8(gray)                 
    return new_img 

img1 = cv2.imread('istockphoto-517188688-612x612.jpg')
img2 = cv2.imread('8EF0FE16-075B-4679-99A6-264B74974F73.jpg')

img1 = cv2.resize(img1, (1000, 500))
img2 = cv2.resize(img2, (1000, 500))

img1 = imgBinQuan(img1[:,:,::-1])
img2 = imgBinQuan(img2[:,:,::-1])

dst = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
plt.imshow(dst)
plt.show()

cv2.imwrite('merge.png', dst)
```
